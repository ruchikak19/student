I really appreciated how smoothly the collaboration went  from brainstorming ideas to dividing tasks and reviewing work.

 What I Worked On: Prompt Engineering

I focused on researching and writing a blog about different types of prompt engineering techniques. These included:

Experience-based prompting  Using learned patterns and previous interactions to refine prompts

Structured prompt design  Crafting prompts with clear instructions, role assignment, and context

Iterative prompting  Improving responses by adjusting and refining prompts step by step

Zero-shot & Few-shot prompting  Getting the AI to perform tasks with little to no examples

The goal was to help others understand how thoughtful prompt design can dramatically improve AI performance, especially in research and digital problem-solving contexts like food insecurity and digital famine studies.

I also focused on integration of drag-and-drop in the code to make an interactive final game experience.
